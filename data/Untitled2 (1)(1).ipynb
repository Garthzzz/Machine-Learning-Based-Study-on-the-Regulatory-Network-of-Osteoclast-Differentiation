```python
import numpy as np
from scipy.stats import ttest_ind

groups = {
    "0h":   [f"RANKL_0h_{i}"   for i in range(1,7)],
    "24h":  [f"RANKL_24h_{i}"  for i in range(1,7)],
    "48h":  [f"RANKL_48h_{i}"  for i in range(1,7)],
    "72h":  [f"RANKL_72h_{i}"  for i in range(1,7)],
}


# ----------------------------------------
# Step 1: Low‐expression filter
# ----------------------------------------
# Keep probes that have mean expression > 1.0 (log2TPM) in at least one induced stage
# (24h, 48h or 72h).
stage_means = {t: df[cols].mean(axis=1) for t, cols in groups.items()}

# Boolean mask: TRUE if probe passes low‐expression threshold in any induced stage
keep_low_expression = (
    (stage_means["24h"] > 1.0) |
    (stage_means["48h"] > 1.0) |
    (stage_means["72h"] > 1.0)
)

df_step1 = df.loc[keep_low_expression].copy()
print(f"After low‐expression filter: {df_step1.shape[0]} probes remain.")
```

```python
# ----------------------------------------
# Step 2: High‑variability filter (CV top 25%)
# ----------------------------------------

# Identify numeric columns (to exclude any non‑numeric metadata)
numeric_cols = df_step1.select_dtypes(include=[np.number]).columns

# Compute mean and standard deviation across the 24 expression samples
means = df_step1[numeric_cols].mean(axis=1)
stds  = df_step1[numeric_cols].std(axis=1)

# Compute coefficient of variation (CV = std / mean)
cv = stds / means

# Determine the 75th percentile cutoff (so that top 35% CV are kept)
cv_cutoff = cv.quantile(0.75)

# Filter probes: keep those with CV >= cutoff
keep_high_variability = cv >= cv_cutoff
df_step2 = df_step1.loc[keep_high_variability].copy()

# Optional: print how many probes remain after filtering
print(f"After high‑variability filter (top 25% CV): {df_step2.shape[0]} probes remain.")
```

```python
# ----------------------------------------
# Step 3: Differential Expression Analysis (DEG)
# ----------------------------------------
# English comments are kept; conversation remains in Chinese.

import numpy as np
import pandas as pd
from scipy.stats import ttest_ind

# Ensure that df_step2 contains only numeric expression columns
# (If there are any stray non‑numeric columns, drop them here.)
numeric_cols = df_step2.select_dtypes(include=[np.number]).columns
df_deg_input = df_step2[numeric_cols].apply(pd.to_numeric, errors='coerce')

# Prepare a counter to record how many comparisons (out of 3) each probe is significant in
sig_counts = pd.Series(0, index=df_deg_input.index)

# Loop over the three induced timepoints
for t in ["24h", "48h", "72h"]:
    # 1) Compute mean expression at 0h and at time t
    mean0 = df_deg_input[groups["0h"]].mean(axis=1)
    meant = df_deg_input[groups[t]].mean(axis=1)
    # 2) Compute log2 fold‑change, adding a tiny constant to avoid division by zero
    log2fc = np.log2((meant + 1e-9) / (mean0 + 1e-9))
    
    # 3) Perform a two‑sample t‑test for each probe
    pvals = []
    for probe in df_deg_input.index:
        vals0 = df_deg_input.loc[probe, groups["0h"]].values.astype(float)
        valst = df_deg_input.loc[probe, groups[t]].values.astype(float)
        # nan_policy='omit' will ignore NaNs if any
        _, p = ttest_ind(vals0, valst, equal_var=False, nan_policy='omit')
        pvals.append(p)
    pvals = pd.Series(pvals, index=df_deg_input.index)
    
    # 4) Identify probes with |log2FC| > 1 AND p < 0.05
    sig = (pvals < 0.05) & (log2fc.abs() > 1.0)
    
    # 5) Increment the significance counter
    sig_counts += sig.astype(int)

# 6) Keep probes significant in at least two of the three comparisons
keep_deg = sig_counts >= 2
df_step3 = df_deg_input.loc[keep_deg].copy()

print(f"After DEG filter (>=2 significant comparisons): {df_step3.shape[0]} probes remain.")

# 7) Save the result
df_step3.to_csv("Filtered_Probes_After_DEG_Filter.csv")
print("Filtered table written to 'Filtered_Probes_After_DEG_Filter.csv'")
```

```python
import numpy as np
from scipy.stats import f_oneway
from statsmodels.stats.multitest import multipletests

# ----------------------------------------
# Step 3: One‑Way ANOVA on df_step2
# ----------------------------------------

# Assume df_step2 already exists in memory from previous steps,
# with rows=probe IDs and cols=24 numeric expression samples.

# Define time‐point groups
groups = {
    "0h":  [f"RANKL_0h_{i}"  for i in range(1,7)],
    "24h": [f"RANKL_24h_{i}" for i in range(1,7)],
    "48h": [f"RANKL_48h_{i}" for i in range(1,7)],
    "72h": [f"RANKL_72h_{i}" for i in range(1,7)],
}

# 1) Collect raw p‐values
probe_list = df_step2.index.tolist()
raw_pvals = []

for probe in probe_list:
    # Extract four groups of replicates
    vals0  = df_step2.loc[probe, groups["0h"]].astype(float).values
    vals24 = df_step2.loc[probe, groups["24h"]].astype(float).values
    vals48 = df_step2.loc[probe, groups["48h"]].astype(float).values
    vals72 = df_step2.loc[probe, groups["72h"]].astype(float).values
    
    # Perform one‐way ANOVA across the four time points
    _, p = f_oneway(vals0, vals24, vals48, vals72)
    raw_pvals.append(p)

raw_pvals = np.array(raw_pvals)

# 2) Adjust p‐values for multiple testing using Benjamini–Hochberg FDR
_, fdr_pvals, _, _ = multipletests(raw_pvals, alpha=0.05, method='fdr_bh')

# 3) Build results DataFrame
anova_df = pd.DataFrame({
    "probe": probe_list,
    "p_value": raw_pvals,
    "p_value_adj": fdr_pvals
}).set_index("probe")

# 4) Filter probes with FDR < 0.05
sig_probes = anova_df[anova_df["p_value_adj"] < 0.05].index
df_anova_filtered = df_step2.loc[sig_probes]

# 5) Output result
print(f"Probes significant by ANOVA (FDR<0.05): {len(sig_probes)}")
df_anova_filtered.to_csv("ANOVA_Significant_Probes.csv")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import math
from scipy.cluster.hierarchy import linkage, fcluster

# ----------------------------------------
# 1. Load your filtered expression matrix
#    (rows=probe IDs, cols= RANKL_0h_1 … RANKL_72h_6)
#    Here we assume df_step2 is already in memory.
#    If not, read it:
# df_step2 = pd.read_csv("Filtered_HighVariability_Matrix.csv", index_col=0)
df = df_step2.copy()

# ----------------------------------------
# 2. Aggregate six replicates per timepoint
# ----------------------------------------
time_groups = {
    "0h":  [f"RANKL_0h_{i}"  for i in range(1,7)],
    "24h": [f"RANKL_24h_{i}" for i in range(1,7)],
    "48h": [f"RANKL_48h_{i}" for i in range(1,7)],
    "72h": [f"RANKL_72h_{i}" for i in range(1,7)],
}
df_avg = pd.DataFrame({
    tp: df[cols].mean(axis=1)
    for tp, cols in time_groups.items()
}, index=df.index)

# ----------------------------------------
# 3. Hierarchical clustering on full 4‐point series
# ----------------------------------------
# Compute the linkage matrix with Ward’s method
X = df_avg.values.astype(float)  # shape = (n_probes, 4)
Z = linkage(X, method="ward")

# Determine a distance threshold automatically by finding the
# largest “jump” in successive linkage distances
distances = Z[:, 2]
sorted_dist = np.sort(distances)
gaps = np.diff(sorted_dist)
idx_max_gap = np.argmax(gaps)
threshold = (sorted_dist[idx_max_gap] + sorted_dist[idx_max_gap + 1]) / 2

# Assign cluster labels based on that threshold
labels = fcluster(Z, t=threshold, criterion="distance")
df_avg["cluster"] = labels

# Report the number of clusters found
n_clusters = len(np.unique(labels))
print(f"Automatically determined cluster count: {n_clusters}")

# ----------------------------------------
# 4. Save probe ID lists for each cluster
# ----------------------------------------
# make sure index is string (if not already)
df_avg.index = df_avg.index.astype(str)

clusters = {
    cid: df_avg.index[labels == cid].tolist()
    for cid in np.unique(labels)
}

with open("auto_clusters.txt", "w") as fout:
    for cid, probes in clusters.items():
        fout.write(f"Cluster {cid} ({len(probes)} probes):\n")
        # map each probe ID to str before joining
        fout.write(", ".join(map(str, probes)) + "\n\n")

# ----------------------------------------
# 5. Randomly select up to 16 clusters & plot their trends
# ----------------------------------------
random.seed(0)

# clusters: dict {cluster_id: [probe_ids, ...], ...}
cluster_ids = list(clusters.keys())
n_sel = min(16, len(cluster_ids))
selected = random.sample(cluster_ids, n_sel)

# determine grid size: 4 columns, as many rows as needed
n_cols = 4
n_rows = math.ceil(n_sel / n_cols)

# make subplots grid
fig, axes = plt.subplots(n_rows, n_cols, figsize=(4*n_cols, 3*n_rows),
                         sharex=True, sharey=True)
axes = axes.flatten()

time_points = [0, 24, 48, 72]

for ax, cid in zip(axes, selected):
    # grab the averaged 4‐point series for this cluster
    subset = df_avg[df_avg["cluster"] == cid].iloc[:, :4]  # 0h~72h
    # plot each probe lightly
    for series in subset.values:
        ax.plot(time_points, series,
                color="gray", alpha=0.3, linewidth=0.8)
    # overlay the mean trajectory
    mean_traj = subset.mean(axis=0).values
    ax.plot(time_points, mean_traj,
            color="red", marker="o", linewidth=2)
    ax.set_title(f"Cluster {cid} (n={len(subset)})", fontsize=8)
    ax.set_xticks(time_points)

# remove unused subplots
for ax in axes[n_sel:]:
    fig.delaxes(ax)

fig.suptitle("Trends for Selected Hierarchical Clusters", fontsize=14)
fig.tight_layout(rect=[0, 0, 1, 0.96])
plt.xlabel("Time (h)")
plt.ylabel("Mean Expression")
plt.show()
```

```python
import pandas as pd
import os

# 1. 读取 ANOVA 过滤后的矩阵，不要把 Probe ID 当成 index
df_anova = pd.read_csv("ANOVA_Significant_Probes.csv", dtype=str)
# 第一列的列名通常是 "0"，我们重命名为 "ProbeID"
first_col = df_anova.columns[0]
df_anova = df_anova.rename(columns={ first_col: "ProbeID" })

# 2. 读取 Platform Table，建立 ProbeID -> GeneSymbol 的映射字典
platform_df = pd.read_csv("Platform_Table.csv", dtype=str)
probe2symbol = platform_df.set_index("ID")["Symbol"].to_dict()


# 4. 将每个簇内部的 ProbeID 列表映射到 Gene Symbol
symbol_clusters = {}
for cid, probe_list in new_clusters.items():
    # 保留只在 df_anova 中出现的 ProbeID
    valid_probes = [p for p in probe_list if p in df_anova["0"].values]
    if not valid_probes:
        continue

    # 映射到 Gene Symbol，并去掉无法映射（None 或 NaN）
    syms = [ probe2symbol.get(p) for p in valid_probes ]
    syms = [ s for s in syms if isinstance(s, str) and s.strip() ]

    if syms:
        symbol_clusters[cid] = syms

print(f"Converted to {len(symbol_clusters)} clusters with Gene Symbols")

# 5. （可选）将结果保存一下，方便后续富集分析
os.makedirs("symbol_clusters", exist_ok=True)
for cid, syms in symbol_clusters.items():
    with open(f"symbol_clusters/cluster_{cid}.txt", "w") as fout:
        fout.write("\n".join(syms))

# 现在，你可以把 symbol_clusters 直接拿去做 gseapy.enrichr() 了，
# 传入 gene_list=symbol_clusters[cid] 就不会再报 int.strip() 的错。

# -------------------------------------------------------------------
# 3. Enrichment analysis: loop over each cluster
# -------------------------------------------------------------------
import os
import time
import gseapy as gp

# clusters: dict of {cluster_id: [gene_symbol, ...], ...}
# symbol_clusters 已经是 {cid: [ "Atg9b", "Cd4", ... ]}

out_root = "enrich"
os.makedirs(out_root, exist_ok=True)

# 只使用 GO Biological Process
gene_sets = ["GO_Biological_Process_2021"]

for cid, full_genes in symbol_clusters.items():
    if not full_genes:
        continue

    # 如果簇太大，只取前 500 个基因
    genes = full_genes[:500] if len(full_genes) > 500 else full_genes

    outdir = os.path.join(out_root, f"cluster_{cid}")
    os.makedirs(outdir, exist_ok=True)

    attempts = 0
    while attempts < 3:
        try:
            enr = gp.enrichr(
                gene_list = genes,
                gene_sets = gene_sets,
                organism  = "Mouse",
                outdir    = outdir,
                cutoff    = 0.05,
                verbose   = False
            )
            enr.run()
            print(f"[OK]   Cluster {cid}: saved → {outdir} (genes={len(genes)})")
            break

        except ValueError as ve:
            # 只有在“no significant terms”时才跳过
            if "No enrich terms" in str(ve):
                print(f"[SKIP] Cluster {cid}: no terms pass cutoff=0.05")
                break
            else:
                # 其它 ValueError 也当做 sending error 处理
                attempts += 1
                print(f"[WARN] Cluster {cid}: ValueError, retry in 5s (#{attempts})")
                time.sleep(5)

        except Exception as ex:
            # 捕获 “Error sending gene list” 或者网络等其它问题
            attempts += 1
            print(f"[WARN] Cluster {cid}: Exception sending list, retry in 5s (#{attempts})")
            time.sleep(5)

    else:
        # 重试 3 次都失败，最终跳过
        print(f"[FAIL] Cluster {cid}: giving up after 3 attempts")

    # 为了保证不过于频繁，哪怕成功也 sleep 一下
    time.sleep(2)

```

```python
import os
import glob
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. 找到所有簇的 Enrichr 报告文件
report_paths = glob.glob("enrich/cluster_*/GO_Biological_Process_2021*.reports.txt")

# 2. 解析所有报告，存到字典里：cluster_id -> DataFrame
enrich_dict = {}
for path in report_paths:
    cid = int(os.path.basename(os.path.dirname(path)).split("_")[-1])
    df = pd.read_csv(path, sep="\t", engine="python")
    # 强制将 Adjusted P-value 转为 float
    df["Adjusted P-value"] = pd.to_numeric(df["Adjusted P-value"], errors="coerce")
    df = df[["Term", "Adjusted P-value"]].dropna(subset=["Adjusted P-value"])
    df["negLog10_adjP"] = -np.log10(df["Adjusted P-value"])
    enrich_dict[cid] = df

print(f"Loaded enrichment for {len(enrich_dict)} clusters.")

# 3. 汇总 TopN 术语，用于热图
top_n = 5
selected_terms = set()
for cid, df in enrich_dict.items():
    # 现在可以正常调用 nsmallest 了
    top_terms = df.nsmallest(top_n, "Adjusted P-value")["Term"].tolist()
    selected_terms.update(top_terms)

heatmap_df = pd.DataFrame(
    index=sorted(selected_terms),
    columns=sorted(enrich_dict.keys()),
    data=np.nan
)

for cid, df in enrich_dict.items():
    sub = df.set_index("Term")
    for term in heatmap_df.index:
        if term in sub.index:
            heatmap_df.at[term, cid] = sub.at[term, "negLog10_adjP"]

# 4. 画热图
plt.figure(figsize=(len(heatmap_df.columns)*0.4+4, len(heatmap_df.index)*0.3+4))
sns.heatmap(
    heatmap_df,
    cmap="Reds",
    linewidths=0.5,
    linecolor="gray",
    cbar_kws={"label": "-log10(Adjusted P-value)"}
)
plt.title("GO BP Enrichment Across Clusters")
plt.xlabel("Cluster ID")
plt.ylabel("GO Term")
plt.tight_layout()
plt.savefig("GO_BP_enrichment_heatmap.png", dpi=500)  # 保存为 PNG，分辨率 300 dpi
plt.show()
```
